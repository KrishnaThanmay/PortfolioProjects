```python
import pandas as pd
import numpy as np

```

```python
df=pd.read_csv('/content/london_merged.csv')
df.head()
```

```python
df.info()
```

```python
#changing column names to our convenience
new_col_names_dict={'timestamp':'date',
                    'cnt':'count',
                    't1':'real_temp_Celsius',
                    't2':'feels_like_temp',
                    'hum':'humidity_percent',
                    'wind_speed':'wind_speed_kph'}
df.rename(new_col_names_dict, axis=1, inplace=True)
```

```python
df['humidity_percent']=df['humidity_percent']/100
#changed humidity values to percentage(i.e. a value between 0 and 1)
```

```python
df.head()
```

```python
df.season.value_counts()# this gives unique values and respective count of them
```

```python
df.weather_code.value_counts()
```

```python
season_dict={'0.0':'spring',
             '1.0':'summer',
             '2.0':'autumn',
             '3.0':'winter'}
weather_dict={'1.0':'clear',
              '2.0':'scattered clouds',
              '3.0':'broken clouds',
              '4.0':'cloudy',
              '7.0':'rain',
              '10.0':'rain with thunderstorm',
              '26.0':'snowfall'}
# firstly the values in season and weather_code columns are integers
#Our first step is to change their data type to string
df.season=df.season.astype(str)
#we use map function to map all those values in given dictionary to our original dataset values
df.season=df.season.map(season_dict)

#Our first step is to change their data type to string
df.weather_code=df.weather_code.astype(str)
#we use map function to map all those values in given dictionary to our original dataset values
df.weather_code=df.weather_code.map(weather_dict)
```

```python
# removing duplicate records
df=df.drop_duplicates()
df.head()
```

```python
#removing any missing values or null values from records
df=df.dropna()
df.head()
```

```python
df.to_excel('london_bike_share_cleaned.xlsx', sheet_name='Data')
```
